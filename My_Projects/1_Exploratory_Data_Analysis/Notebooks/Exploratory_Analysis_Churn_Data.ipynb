```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Dataset
file_location = r'C:\Users\T992205\why_so_serious\My_Projects\1_Exploratory_Data_Analysis\Data\WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(file_location)
```

```python
# Exploratory Data Analysis
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.show()
```

```python
# Distribution of Monthly Charges
sns.histplot(data=df, x='MonthlyCharges', bins=30)
plt.title('Distribution of Monthly Charges')
plt.show()
```

```python
sns.pairplot(df[['MonthlyCharges', 'TotalCharges', 'tenure']])
plt.title('Pairplot of Numeric Features')
plt.show()
```

```python
sns.countplot(data=df, x='gender', hue='Churn')
plt.title('Churn by Gender')
plt.show()
```
